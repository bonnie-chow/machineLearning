```python
def check_if_number_of_appearances_are_below_threshold(name, threshold, dictionary):
        for element, number_appearances in dictionary.items():
            if element == name:
                if number_appearances < threshold:
                    return True
                return False
        return False


check_if_number_of_appearances_are_below_threshold('Harrison Ford', 3, main_actors_with_number_of_appearances)
```

```python
from collections import Counter

train['main_character'] = train['main_character'].apply(lambda x: 'other' if check_if_number_of_appearances_are_below_threshold(x, 3, main_actors_with_number_of_appearances) else x)
```
